```python
from sympy import *
init_printing(use_unicode=True, wrap_line=False)
u, v, ti, tj, tip1, tjp1 = symbols('u,v,ti,tj,tip1,tjp1', positive=True)
sigma1, sigma2 = symbols('sigma1,sigma2', positive=True)
dti, dtj, wi, wj = symbols('dti,dtj,wi,wj', positive=True)
```

```python
wu = (1-exp(-sigma1*(u-ti)))
wv = (1-exp(-sigma2*(v-tj)))
first_integral = simplify(integrate(wu * (u-v), (u, ti, tip1)))
print(first_integral)
```

```python
vpart = (sigma1**2*(+ 2*ti*v -2*tip1*v) + 2*sigma1*v - 2 * sigma1*v*exp(sigma1*(ti - tip1)))/(2*sigma1**2)
rest = (sigma1**2*(-ti**2 + tip1**2) - 2*sigma1*ti + 2*(sigma1*tip1 + 1)*exp(sigma1*(ti - tip1)) - 2)/(2*sigma1**2)
vpart = simplify(vpart)
print(vpart)
print(rest)
```

```python
# check that we have separated it correctly
simplify(first_integral - vpart - rest)
```

### Now complete the rest of the integration split for simplification

```python
int_vpart = vpart / v * simplify(integrate(v * wv, (v, tj, tjp1)))
int_vpart
```

```python
int_rest = rest * integrate(wv, (v, tj, tjp1))
int_rest
```
